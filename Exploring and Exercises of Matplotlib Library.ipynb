```python
#Marker Size
#You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

import numpy as np
ypoints=np.array([4,6,2,8,4])

plt.plot(ypoints,'*--b',ms=30)#Setting marker size to 20

plt.show()
```

```python
#Marker edge Color
#You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:
import numpy as np
ypoints=np.array([4,6,2,8,4])

plt.plot(ypoints,'*--g',ms=30,mec='b')#Setting marker edge colour blue

plt.show()
```

```python
#Marker face Color
#You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:
#You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:
#Use both the mec and mfc arguments to color the entire marker:
import numpy as np
ypoints=np.array([4,6,2,8,4])

plt.plot(ypoints,'*--g',ms=30,mec='b',mfc='m')#Setting marker edge colour blue

plt.show()
```

```python
#You can also use Hexadecimal color values:
import numpy as np
ypoints=np.array([4,6,2,8,4])

plt.plot(ypoints,'d-b',ms=15,mec='#16bf0a',mfc='#166204')#Setting marker edge colour blue

plt.show()
```

```python
#Create Labels for a Plot
#With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.
import matplotlib.pyplot as plt
import numpy as np
xaxis=np.array([10,32,45,65,78,90,100,190,400,270,500])
yaxis=np.array([12,34,66,773,36,200,88,99,34,56,400])
plt.plot(xaxis,yaxis)
plt.title("Data About Pulse and Calories burned")#With Pyplot, you can use the title() function to set a title for the plot.

plt.xlabel("calories burned")
plt.ylabel("average pulse")
plt.show()
```

```python
#Set font properties for the title and labels:
import matplotlib.pyplot as plt
import numpy as np

xaxis=np.array([10,32,45,65,78,90,100,190,400,270,500])
yaxis=np.array([12,34,66,773,36,200,88,99,34,56,400])
plt.plot(xaxis,yaxis)

"""Common serif fonts include Times New Roman, Georgia, and Times. 
When you specify 'serif' as the font family in your dictionary, 
you are telling a library like Matplotlib (used for creating plots and graphs) to use a serif font for rendering text,
 and you can further customize its appearance using other properties in the dictionary, such as 'color' and 'size'."""

""" use the fontdict parameter in xlabel(), ylabel(), and title() 
to set font properties for the title and labels."""

fontdict1={"family":'serif','color':'blue','size':20}#here family means fontfamily serif is a font of family
fontdict2={"family":'serif','color':'green','size':20}

plt.title("Sports data ",fontdict=fontdict1,loc='left')#loc parameter in title() to position the title.

plt.xlabel("calories burned",fontdict=fontdict2)
plt.ylabel("average pulse",fontdict=fontdict2)

"""axis parameter in the grid() function to specify which grid lines to display.

Legal values are: 'x', 'y', and 'both'. Default value is 'both'."""

"""line properties of the grid, like this: 
grid(color = 'color', linestyle = 'linestyle', linewidth = number)."""

plt.grid(color='Blue',linestyle='--',linewidth=0.5) #grid() function to add grid lines to the plot.


#plt.grid(axis='x')#axis parameter specifying the grid line wants only for x-axis

plt.show()
```

```python
#Subplot function
#You can add a title to each plot with the title() function:
#add a title to the entire figure with the suptitle() function:

import matplotlib.pyplot as plt
import numpy as np
arr=np.array([0,1,2,3])
arr2=np.array([6,8,9,56])

#Plot1#
plt.subplot(1,2,1)##the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(arr,arr2)
plt.title('Subplot1')#adding title withe title()

x=np.array([0,1,2,3])
y=np.array([5,8,9,22])

#plot2
plt.subplot(1,2,2)##the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)
plt.title('Subplot2')

plt.suptitle("Sample Subplots")#add a title to the entire figure with the suptitle() function:
plt.show()
```

```python
#Color Each Dot
#You cannot use the color argument for this, only the c argument.
import matplotlib.pyplot as plt
import numpy as np

x=np.array([4,6,3,6,4,4,6,8,9,4,7])
y=np.array([4,6,3,5,4,45,6,8,63,73,5])

colors=np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray"])

plt.scatter(x,y,c=colors)#c argument used

plt.show()
```

```python
#ColorMap
#viridis
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

#A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])#Create a color array, and specify a colormap in the scatter plot:

"""you can specify the colormap with the keyword argument cmap with the value of the colormap,
 in this 'viridis' which is one of the built-in colormaps available in Matplotlib.

In addition you have to create an array with values (from 0 to 100), one value for each point in the scatter plot:"""

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()#You can include the colormap in the drawing by including the plt.colorbar() statement

plt.show()
```
